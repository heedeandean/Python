```python
z = np.linspace(-1, 1, 100)
w = np.sqrt(1-z**2)
xo = np.concatenate([z,z[::-1][1:100]])
yo = np.concatenate([w,-1*w[::-1][1:100]])
xo
```

```python
fig, axs = plt.subplots(2, 2, sharex='col', sharey='row', # 공통축
                        gridspec_kw={'hspace': 0, 'wspace': 0})

(ax1, ax2), (ax3, ax4) = axs
fig.suptitle('Sharing x per column, y per row')
ax1.plot(xo-1, yo)
ax2.plot(xo-1, yo-1, 'orange')
ax3.plot(xo+1, yo+1, 'green')
ax4.plot(xo+1, yo, 'red')
```

```python
import scipy.stats as stats

xh = np.random.randn(4000)
xh
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(xh, bins=30, color='skyblue', rwidth=0.9, label='Empirical')
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xs = np.linspace(xlim[0], xlim[1], 200)
dx = stats.norm.pdf(xs)
dx = dx*(ylim[1]/dx.max())
plt.plot(xs, dx, 'r-', label='PDF', color='orange')
ax.set_ylim((ylim[0], 1.2*ylim[1]))
plt.legend()
plt.text(1, 320, r'normal density: $\phi(x)$')
plt.title('Histogram and Density')
```

```python
sns.set(style='darkgrid')
iris = sns.load_dataset('iris')
iris.__class__
```

```python
iris.columns
```

```python
iris.columns = ['sl', 'sw', 'pl', 'pw', 'sp']
iris.head()
```

```python
# 1. histogram
g = sns.FacetGrid(iris, col='sp')
g = g.map(plt.hist, 'sl')

# 2. boxplot
sns.catplot(x='sp', y='sl', kind='boxen', data=iris)

# 3. 1, 2차원 커널밀도함수
sns.jointplot(x='sw', y='sl', kind='kde', data=iris, color='g')

# 4. 짝별 산점도: 모든 변수 간 관계 파악 가능
# 대각선: 1차원 커널밀도함수
# 대각선 아래: 2차원 커널밀도함수
g = sns.PairGrid(iris, hue='sp')
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
g.map_diag(sns.kdeplot, lw=3, legend=False)

# 5. 회귀
# 심슨의 역설: 두 변수 x와 y의 관계가 제3의 통제변수('sp')를 고려할 때와 그렇지 않은 때의 해석이 정반대로 달라짐.
sns.lmplot(x='sw', y='sl', data=iris)
sns.lmplot(x='sw', y='sl', data=iris, hue='sp')
```

```python
import pandas as pd

url = 'https://media.githubusercontent.com/media/neurospin/pystatsml/master/datasets/salary_table.csv'
salary_df = pd.read_csv(url)
salary_df.head()
```

```python
colors_edu = {'Bachelor':'r', 'Master':'g', 'Ph.D':'blue'}
plt.scatter(salary_df['experience'], salary_df['salary'],
            c=salary_df['education'].apply(lambda x:colors_edu[x]), s=100)
plt.show()
```

```python
plt.ion()

symbols_manag = dict(Y='*', N='.')
for values, d in salary_df.groupby(['education', 'management']):
    edu, manager = values
    plt.scatter(d['experience'], d['salary'],
                marker=symbols_manag[manager], color=colors_edu[edu],
                s=150, label=manager+'/'+edu)
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.legend(loc=4) # lower right
```
