```python
# 병합
df1 = pd.DataFrame({'A': ['A0', 'A1'], 'B': ['B0', 'B1']})
df2 = pd.DataFrame({'A': ['A2', 'A3'], 'B': ['B2', 'B3']})
pd.concat([df1, df2], ignore_index=True) # 행(레코드) 추가 (기본값)
```

```python
pd.concat([df1, df2], axis=1, ignore_index=False) # 열(속성) 추가
```

```python
left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'], 'A': ['A0', 'A1', 'A2', 'A3']})
right = pd.DataFrame({'key': ['K1', 'K2', 'K3', 'K4'], 'B': ['B0', 'B1', 'B2', 'B3']})
pd.merge(left, right, on='key', how='inner') # default(교집합)
```

```python
pd.merge(left, right, on='key', how='outer') # 합집합
```

```python
pd.merge(left, right, on='key', how='left')
```

```python
pd.merge(left, right, on='key', how='right')
```
