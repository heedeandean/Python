```python
price_sort_index = price.sort_index(ascending=False) # 인덱스 기준 정렬
price_sort_index.head()
```

```python
price_sort_values = price.sort_values(by='close', ascending=False) # 열 기준 정렬
price_sort_values.head()
```

```python
price['month'] = price.index.map(
    lambda date: date.strftime('%Y') + '-' + date.strftime('%m') # datetime -> 문자열 타입
)
price.head()
```

```python
# 그룹별 집계
price_mean = (price.groupby(by='month')[['open', 'close']]
              .agg(open_mean=('open', 'mean'),
                   close_mean=('close', 'mean'))
              .reset_index())
price_mean.head()
```

```python
price_mean = (price.groupby(by='month')
              .agg({'open': 'mean', 'close': 'mean'})
              .reset_index())
price_mean.head()
```

```python
# 열이름 변경
price_mean.rename(
    {'open': 'open_mean', 'close': 'close_mean'},
    axis=1,
    inplace=True
)
price_mean.head()
```
